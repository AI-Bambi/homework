■問題
以下の二次関数のグラフをMatplotlibで描いてみましょう。

y = x^2 - 4x + 3 という二次関数を0から5までの範囲でプロットする。

目盛線を追加し、色と線のスタイルを変更する。

```python
import numpy as np

import matplotlib.pyplot as plt

x=np.arange(0, 6,1)
y=x**2 - 4*x +3

plt.plot(x, y,label='$y = x^2 - 4x + 3$', color='red', linestyle='-.',linewidth=2.0) #label=r'$y = x^2 - 4x + 3$'

plt.grid(True, which='both', axis='both', linestyle='-', color='gray', alpha=0.3)

#plt.axhline(0, color='black',linewidth=1)
#plt.axvline(0, color='black',linewidth=1)

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
