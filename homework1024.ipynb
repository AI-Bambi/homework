```python
import random
import numpy
import pandas as pd
import matplotlib.pyplot as plt

class Product:
    def __init__(self):
        self.goods_name=""
        self.price=0
        self.color=""
        self.stock=0

    def calc_sell(self): #同じdef下と同じ名前は付けない
        sell_amount=random.randint(0,self.stock)
        #print(sell_amount)
        self.sales=sell_amount * self.price


class SetProduct:
    def __init__(self,product1,product2):
        self.product1=product1
        self.product2=product2

    def set_price(self):
      set_total_price=self.product1.price + self.product2.price
      return set_total_price - 100

    def sell(self, quantity):
        total_quantity = min(self.product1.stock, self.product2.stock, quantity)
        self.product1.stock -= total_quantity
        self.product2.stock -= total_quantity
        return self.set_price() * total_quantity

ball=Product()
glove=Product()
bat=Product()
uniform=Product()

ball.stock=20
glove.stock=20
bat.stock=20
uniform.stock=20

ball.price=500
glove.price=1000
bat.price=2000
uniform.price=4000



#それぞれの製品の売り上げ数×単価をかける
ball.calc_sell()
glove.calc_sell()
bat.calc_sell()
uniform.calc_sell()

ball.goods_name="ボール"
glove.goods_name="グローブ"
bat.goods_name="バット"
uniform.goods_name="ユニフォーム"

salses_check=[ball.sales,glove.sales,bat.sales,uniform.sales]
goods_name=["ボール","グローブ","バット","ユニフォーム"]
max_color_number=numpy.argmax(salses_check)
max_color=goods_name[max_color_number]

products = [ball, glove, bat, uniform]
set_product1 = SetProduct(products[0], products[1])  # ボールとグローブのセット
set_product2 = SetProduct(products[2], products[3])

# セット商品の売上を計算
set_quantity_sold1 = random.randint(0, min(set_product1.product1.stock, set_product1.product2.stock))
set_sales_amount1 = set_product1.sell(set_quantity_sold1)

set_quantity_sold2 = random.randint(0, min(set_product2.product1.stock, set_product2.product2.stock))
set_sales_amount2 = set_product2.sell(set_quantity_sold2)


print(f"今日の{ball.goods_name}の売上は{ball.sales}円です。")
print(f"今日の{glove.goods_name}の売上は{glove.sales}円です。")
print(f"今日の{bat.goods_name}の売上は{bat.sales}円です。")
print(f"今日の{uniform.goods_name}の売上は{uniform.sales}円です。")
print(f"ボールとグローブのセットの売上は{set_sales_amount1}円です。")
print(f"バットとユニフォームのセットの売上は{set_sales_amount2}円です。")
#print(max_color)
print(f"一番売上が大きかった商品は{max_color}で{max(salses_check)}円です。")

labels = goods_name + [f"{set_product1.product1.goods_name} & {set_product1.product2.goods_name}",
                       f"{set_product2.product1.goods_name} & {set_product2.product2.goods_name}"]
sales_values = salses_check + [set_sales_amount1, set_sales_amount2]

plt.bar(labels, sales_values, color=['blue', 'orange', 'green', 'red', 'purple', 'cyan'])
plt.title('商品別売上')
plt.xlabel('商品名')
plt.ylabel('売上金額 (円)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
